# PREPARING THE DATASET


```python
happii_2023.loc[happii_2023.duplicated()] # we check duplicated rows
```

```python
# Remove all columns between column name 'Ladder score in Dystopia' to 'Dystopia + residual' 
happii_2023 = happii_2023.drop(happii_2023.loc[:, 'Ladder score in Dystopia':'Dystopia + residual'].columns, axis=1)
# Remove all columns between column name 'Standard error of ladder score' to 'lowerwhisker' 
happii_2023 = happii_2023.drop(happii_2023.loc[:, 'Standard error of ladder score':'lowerwhisker'].columns, axis=1)
```

```python
#Create integer rank column 
happii_2023['rank'] = happii_2023['Ladder score'].rank(ascending=False)
happii_2023['rank'] = happii_2023['rank'].astype(int)
```

```python
#Rename the columns for consistency
happii_df_2023 = happii_2023.rename({'Country name':'country','Standard error of ladder score':'standard_error_of_ladder_score'
                                   , 'Ladder score':'happiness_score','Happiness score':'happiness_score'
                                   , 'Logged GDP per capita':'gdp_per_capita','Social support':'social_support'
                                   , 'Healthy life expectancy':'healthy_life_expectancy'
                                   , 'Freedom to make life choices':'freedom_to_make_life_choices'
                                   , 'Generosity':'generosity','Perceptions of corruption':'perceptions_of_corruption'
                                   , 'Explained by: Freedom to make life choices':'freedom_to_make_life_choices'
                                   , 'Explained by: Generosity':'generosity'
                                   , 'Explained by: Perceptions of corruption':'perceptions_of_corruption'}, axis =1)
happii_df_2023.head()
```

```python
#Create top 10 and bottom 10 identifier column

def top_bottom_identifier(value):
    if value < 11:
        return "top 10 happiest"
    if value > 127:
        return "bottom 10 happiest"
    elif 11 <= value < 128:
        return "not top/bottom 10"
    
 
happii_df_2023['top_bottom_identifier'] = happii_df_2023['rank'].map(top_bottom_identifier)
```

```python
country_mapping.head()
```

```python
# Dropping irrelevant columns
country_mapping.drop('alpha-2', inplace=True, axis=1)
# Remove all columns between column name 'country-code' to 'iso_3166-2' 
country_mapping = country_mapping.drop(country_mapping.loc[:, 'country-code':'iso_3166-2'].columns, axis=1)
# Remove all columns between column name 'intermediate-region' to 'intermediate-region-code' 
country_mapping = country_mapping.drop(country_mapping.loc[:, 'intermediate-region':'intermediate-region-code'].columns, axis=1)
country_mapping.head()
```

```python
#Rename the columns for consistency
country_mapping = country_mapping.rename({'name':'country','alpha-3':'iso_alpha','sub-region':'sub_region'}, axis =1)
```

```python
#Merge the happiness data and country mapping dataframes
happii_region_df = happii_df_2023.merge(country_mapping, on='country', how='left')
happii_region_df
```

```python
#Check for nulls
happii_region_df.isnull().sum()
```

```python
# Indentifying the null 'healthy_life_expectancy' row
nan_hlf_rows = happii_region_df[happii_region_df['healthy_life_expectancy'].isnull()]
nan_hlf_rows
```

```python
# Check for rows with null region that did not match correctly when merging
nan_region_rows = happii_region_df[happii_region_df['region'].isnull()]
nan_region_rows
```

```python
#Lets change the names in the country_mapping df to match those of the happiness report where appropriate

#Turkey ----> Turkiye
country_mapping['country'] = country_mapping['country'].str.replace('Turkey', 'Turkiye', regex=True)
#Palestine, State of ----> State of Palestine
country_mapping['country'] = country_mapping['country'].str.replace('Palestine, State of', 'State of Palestine', regex=True)
#Côte D'Ivoire ----> Ivory Coast
country_mapping['country'] = country_mapping['country'].str.replace("Côte D'Ivoire", 'Ivory Coast', regex=True)
#Macedonia ----> North Macedonia
country_mapping['country'] = country_mapping['country'].str.replace('Macedonia', 'North Macedonia', regex=True)
#Hong Kong ----> Hong Kong S.A.R. of China
country_mapping['country'] = country_mapping['country'].str.replace('Hong Kong', 'Hong Kong S.A.R. of China', regex=True)
#Taiwan ----> Taiwan Province of China
country_mapping['country'] = country_mapping['country'].str.replace('Taiwan', 'Taiwan Province of China', regex=True)
#Czech Republic ----> Czechia
country_mapping['country'] = country_mapping['country'].str.replace('Czech Republic', 'Czechia', regex=True)
```

```python
#Merge the dataframes again
happii_region_df = happii_df_2023.merge(country_mapping, on='country', how='left')
```

```python
#Check for nulls
happii_region_df.isnull().sum()
```

```python
#Lets fix the nulls in the last four rows manually
nan_region_rows = happii_region_df[happii_region_df['region'].isnull()]
nan_region_rows
```

```python
#Manually updating region, sub_region and iso code for Kosovo
happii_region_df.loc[33,'region'] = 'Europe'
happii_region_df.loc[33,'sub_region'] = 'Southern Europe'
happii_region_df.loc[33,'iso_alpha'] = 'XXK'

#Manually updating region, sub_region and iso code for Bosnia and Herzegovina
happii_region_df.loc[70,'region'] = 'Europe'
happii_region_df.loc[70,'sub_region'] = 'Southern Europe'
happii_region_df.loc[70,'iso_alpha'] = 'BIH'

#Manually updating region, sub_region and iso code for Congo (Brazzaville)
happii_region_df.loc[85,'region'] = 'Africa'
happii_region_df.loc[85,'sub_region'] = 'Sub-Saharan Africa'
happii_region_df.loc[85,'iso_alpha'] = 'COG'

#Manually updating region, sub_region and iso code for Congo (Kinshasa)
happii_region_df.loc[132,'region'] = 'Africa'
happii_region_df.loc[132,'sub_region'] = 'Sub-Saharan Africa'
happii_region_df.loc[132,'iso_alpha'] = 'COD'
```

# UNDERSTANDING FEATURES


```python
#Creating top 10 and bottom 10 data frames and concatinating them
top10 = happii_region_df.set_index('country')['happiness_score'].nlargest(10).to_frame()
bottom10 = happii_region_df.set_index('country')['happiness_score'].nsmallest(10).to_frame()
dfconcat = pd.concat([top10, bottom10], axis=0)
dfconcat = dfconcat.sort_values(by="happiness_score")

fig = plt.figure(figsize=(10,8))
cmap = plt.cm.RdYlGn
norm = plt.Normalize(vmin=dfconcat['happiness_score'].min(), vmax=dfconcat['happiness_score'].max())
colors = [cmap(norm(value)) for value in dfconcat['happiness_score']]

g = plt.barh(y=dfconcat.index, width=dfconcat["happiness_score"], color=colors)
plt.ylabel("Country")
plt.xlabel("Happiness Score")
plt.title("10 Happiest and 10 Least Happiest Countries", fontsize = 30)

plt.show()
```

```python
# Look at the distribution of happiness scores, plotting mean (red) and median (black) lines

plt.figure(figsize=(10,6))
plt.title("Happiness Score Distribution of the Countries , 2023", fontsize = 15)
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.hist(happii_region_df.happiness_score, bins = np.arange(0,10,0.5), edgecolor = 'black')

plt.axvline(x = happii_region_df['happiness_score'].mean(), color = 'red')
plt.axvline(x = happii_region_df['happiness_score'].median(), color = 'green')
```

```python
#Let's create kernel density estimates for the 6 factors

fig = plt. figure()
fig, axes = plt.subplots(2, 3, figsize= (10, 6))
plt.tight_layout(pad= 3)


plta = sns.kdeplot( data=happii_region_df, x='gdp_per_capita',color='green',fill=True, ax=axes[0,0])
pltb = sns.kdeplot( data=happii_region_df, x='healthy_life_expectancy',color='green',fill=True, ax=axes[0,1])
pltc = sns.kdeplot( data=happii_region_df, x='social_support',color='green',fill=True, ax=axes[0,2])
pltd = sns.kdeplot( data=happii_region_df, x='freedom_to_make_life_choices',color='green',fill=True, ax=axes[1,0])
plte = sns.kdeplot( data=happii_region_df, x='generosity',color='green',fill=True, ax=axes[1,1])
pltf = sns.kdeplot( data=happii_region_df, x='perceptions_of_corruption',color='green',fill=True, ax=axes[1,2])


plta.axvline(x = happii_region_df['gdp_per_capita'].mean(), color = 'red')
plta.axvline(x = happii_region_df['gdp_per_capita'].median(), color = 'black')
pltb.axvline(x = happii_region_df['healthy_life_expectancy'].mean(), color = 'red')
pltb.axvline(x = happii_region_df['healthy_life_expectancy'].median(), color = 'black')
pltc.axvline(x = happii_region_df['social_support'].mean(), color = 'red')
pltc.axvline(x = happii_region_df['social_support'].median(), color = 'black')
pltd.axvline(x = happii_region_df['freedom_to_make_life_choices'].mean(), color = 'red')
pltd.axvline(x = happii_region_df['freedom_to_make_life_choices'].median(), color = 'black')
plte.axvline(x = happii_region_df['generosity'].mean(), color = 'red')
plte.axvline(x = happii_region_df['generosity'].median(), color = 'black')
pltf.axvline(x = happii_region_df['perceptions_of_corruption'].mean(), color = 'red')
pltf.axvline(x = happii_region_df['perceptions_of_corruption'].median(), color = 'black')
pltf.axvline(x = happii_region_df['perceptions_of_corruption'].median(), color = 'black')
```
